# Importación de librerías

```python
import cv2
from google.colab.patches import cv2_imshow
```

# Lectura y visualización de la imagen
Se carga la imagen en formato BGR, se muestra su tamaño, tipo de dato y memoria utilizada. Luego, se visualiza.

```python
imagen_bgr = cv2.imread('/content/perrito.jpg')
print("Dimensiones de la imagen:", imagen_bgr.shape)
print("Tipo de datos:", imagen_bgr.dtype)
print("Tamaño en memoria (bytes):", imagen_bgr.nbytes)
cv2_imshow(imagen_bgr)
```

# Canal Azul (B)
Se extrae el canal azul de la imagen (índice 0), se calculan estadísticas básicas y se muestra el canal.

```python
B = imagen_bgr[:,:,0]
print("Estadísticas del canal Azul:")
print(f"Valor mínimo: {B.min()}, Valor máximo: {B.max()}, Promedio: {B.mean():.2f}")
cv2_imshow(B)
```

# Canal Verde (G)
Se extrae el canal verde (índice 1), se muestran estadísticas y su visualización.

```python
G = imagen_bgr[:,:,1]
print("\nEstadísticas del canal Verde:")
print(f"Valor mínimo: {G.min()}, Valor máximo: {G.max()}, Promedio: {G.mean():.2f}")
cv2_imshow(G)
```

# Canal Rojo (R)
Se analiza el canal rojo (índice 2) y se presentan sus características.

```python
R = imagen_bgr[:,:,2]
print("\nEstadísticas del canal Rojo:")
print(f"Valor mínimo: {R.min()}, Valor máximo: {R.max()}, Promedio: {R.mean():.2f}")
cv2_imshow(R)
```

# Importación de Matplotlib
Se importa la biblioteca matplotlib.pyplot para una visualización alternativa.

```python
import matplotlib.pyplot as plt
```

# Visualización con Matplotlib
Se carga la imagen en formato RGB y se muestra con Matplotlib, incluyendo título, ejes y barra de color.

```python
imagen_rgb = plt.imread('/content/perrito.jpg')#carga la imagen como una matriz NumPy, donde cada píxel tiene valores entre 0 y 1 (si es RGB normalizado).
plt.figure(figsize=(10,8))
plt.imshow(imagen_rgb)#interpreta esos valores y los representa visualmente.
plt.title('Imagen RGB en Matplotlib')
plt.axis('on')
plt.colorbar()#agrega una escala lateral que muestra cómo se relacionan los valores numéricos de los píxeles con los colores.
plt.show()
```
